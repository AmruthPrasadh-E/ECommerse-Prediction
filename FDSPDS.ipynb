## LINEAR REGRESSION MODEL:

-Simple linear regression analysis between features "Yearly Amount Spent" and "Length of Membership".

-"Length of Membership" is treated as independent variable(ie. Feature), and "Yearly Amount Spent" is treated as 
  dependent variable.

```python
from sklearn.linear_model import LinearRegression
#Import LinearRegression from sklearn.linear_model
```

```python
X1 = df['Length of Membership'].values
print("X1:\n",X1)
Y1 = df['Yearly Amount Spent'].values
print("Y1:\n",Y1)
```

```python
X1=X1.reshape(len(X1),1)
reg = LinearRegression().fit(X1, Y1)
```

```python
plt.scatter(X1, Y1, label='Data points')
```

```python
Y_pred = reg.predict(X1)
print(Y_pred)
```

```python
r2_square = reg.score(X1, Y1)
print("R-squared value:\t",r2_square)
```

-R-Squared is a statistical measure of fit that indicates how much variation of a dependent variable is explained by the independent variable(s) in a regression model.

-R-squared values range from 0 to 1 and are commonly stated as percentages from 0% to 100%. An R-squared of 100% means that all movements of a security (or another dependent variable) are completely explained by movements in the index (or the independent variable(s) you are interested in).

## Inferences from the linear reggression model:

- The linear regression model shows linear relationship of dependent feature "Yearly Amount Spent" and independent feature "Length of Membership". 

- From the R-squared value we can infer that the Regression model is considered relatively strong with the value of 0.654.

# Multiple Regression model:

- Regression model built from more than one variable.

For building the multiple regression model consider the following procedures.

## Training and Testing Data


```python
X2=df[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership' ]]
```

```python
Y2=df['Yearly Amount Spent']
```

## Spliting the data into training and testing sets. 

```python
from sklearn.model_selection import train_test_split
# train_test_split function - used to split the dataset into training and testing sets.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, Y2, test_size=0.3, random_state=101)
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
#Import LinearRegression from sklearn.linear_model
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
coeff_df=pd.DataFrame(reg.coef_,X2.columns,columns=['Coefficients'])
coeff_df   #the coefficients of the model
```

## Prediction of Test Data

```python
predictions=reg.predict(X_test)      #reg.predict()- to predict off the X_test set of the data.
```

## Real test values versus the predicted values.

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y test')
plt.ylabel('Y_predicted')
```

```python
import pickle
```

```python
pickle.dump(reg,open("model.pkl","wb"))
```

```python
reg.predict(pd.DataFrame(columns=X2.columns, data=np.array([["34.497268","12.655651","39.577668","4.082621"]])))
```

## Evaluating the Model:

Calculating the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

##  Adjusted R-Squared value:

- The adjusted R-squared compares the descriptive power of regression models that include diverse numbers of predictors.

- Every predictor added to a model increases R-squared and never decreases it. Thus, a model with more terms may seem to have a better fit.

```python
sns.histplot((y_test-predictions),bins=50,kde=True);
```

## Inferences from Multiple Regression model:

- The pregictions made on dependent variable "Yearly Amount Spent" shows comparitively better results than that of Simple regression model

- Normal distribution shows the probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.

- It shows that the model selection was accurate.
